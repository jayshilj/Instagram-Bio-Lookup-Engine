## Function to return bio of a user from private key

```python
def extract_bio(user_pk) :
    api.getUsernameInfo(user_pk)
    temp=api.LastJson['user']
    bio=temp['biography']
    if(len(bio)==0):
        print("-")
        bio="none"
    else:
        print('ok')
    return bio
```

## Let's generate 20 random profile information

```python
users = [8538970820,4550727270,295336107,1359075101,27985392,1122362258,23081249,288804792,236134296,300582150,
           2380119692,4610812401,214899198,232652088,4831050466,230965921,1477174194,392838699,5699575544,301517011]
```

```python
data=[]
```

```python
for x in users:
    api.getUserFollowings(x)
    following=api.LastJson
    tempdf=pd.DataFrame(following['users'])
    data.append(tempdf)
```

```python
len(data)
```

```python
print(data)
```

```python
frames = [data[i] for i in range(len(data))]
total_df = pd.concat(frames)
```

```python
start = 0
count_until_sleep = 90
x=0

sleep_at_count = start+count_until_sleep
for i in range(0,total_df.shape[0]):
    pk=total_df.iloc[i]['pk']
    api.getUsernameInfo(pk)
    temp=api.LastJson['user']
    x=x+1
    print("Able to get user number ", x)
    bio=temp['biography']
    if(len(bio)==0):
        bio='none'
    total_df.loc[total_df['pk']==pk, 'biography']=bio
    sleep(0.1)
    if(i==sleep_at_count):
        sleep_at_count+=count_until_sleep
        print(total_df[i-1:i+1])
        total_df.to_csv('temporary_instagram', index=False)
        print("Stopping for 30 minutes at time:")
        print(datetime.datetime.now())
        sleep(1800)
        print("resuming at: ",x)
        print(datetime.datetime.now())
```
